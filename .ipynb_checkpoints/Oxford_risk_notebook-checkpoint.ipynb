#### Prepare feature matrix & common scaler

```python
# Code 4-Setup
from sklearn.preprocessing import StandardScaler
import numpy as np

FEATURES_TRAITS = ['confidence', 'risk_tolerance',
                   'composure', 'impulsivity', 'impact_desire']

FEATURES_PORT   = ['portfolio_value', 'n_assets',
                   'mean_value', 'n_currencies', 'n_alloc_types']

# OPTION 1 – traits only
X_traits = merged[FEATURES_TRAITS].values
X_traits_scaled = StandardScaler().fit_transform(X_traits)

# OPTION 2 – traits + portfolio
X_all = merged[FEATURES_TRAITS + FEATURES_PORT].values
X_all_scaled = StandardScaler().fit_transform(X_all)
```



### Principal Component Analysis (PCA)
#### How many latent factors describe investor traits?

```python
# Code 4-A – PCA on trait space
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA().fit(X_traits_scaled)
expl_var = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(5,3))
plt.plot(range(1, len(expl_var)+1), expl_var, marker='o')
plt.ylabel('Cumulative explained variance')
plt.xlabel('# Principal components')
plt.title('Trait PCA – variance explained')
plt.grid(True)
plt.show()

print("Explained variance per component:",
      pca.explained_variance_ratio_.round(3))
```

#### Visualising first two PCs coloured by risk quartile

```python
# Code 4-B
pc_scores = pca.transform(X_traits_scaled)[:, :2]
merged[['PC1_traits','PC2_traits']] = pc_scores

sns.scatterplot(data=merged, x='PC1_traits', y='PC2_traits',
                hue='risk_quartile', palette='viridis')
plt.title('Investors in trait-PCA space')
plt.show()
```


### Non-linear embedding (t-SNE)
#### t-SNE projection of traits+portfolio metrics

```python
# Code 4-C 
from sklearn.manifold import TSNE

tsne = TSNE(init='random', learning_rate='auto',
            perplexity=30, random_state=42)
tsne_emb = tsne.fit_transform(X_all_scaled)

merged[['TSNE1','TSNE2']] = tsne_emb

sns.scatterplot(data=merged, x='TSNE1', y='TSNE2',
                hue='risk_quartile', palette='Spectral')
plt.title('t-SNE of investors (traits + portfolio)')
plt.show()
```

### Clustering investors

#### K-means on combined feature space: how many clusters?

```python
# Code 4-D
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silh = {}
for k in range(2, 8):
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)
    labels = km.fit_predict(X_all_scaled)
    silh[k] = silhouette_score(X_all_scaled, labels)

plt.plot(list(silh.keys()), list(silh.values()), marker='o')
plt.title('Silhouette score vs K')
plt.xlabel('K')
plt.ylabel('Score')
plt.show()

best_k = max(silh, key=silh.get)
print("Best K by silhouette:", best_k)
```

Outcome:

1. We explored how many natural investor segments emerge when K-means is applied to the combined trait + portfolio feature space.

2. Silhouette analysis peaks at K = 4 with a score of ~ 0.167.

K = 2 already shows some separability (0.165), but adding clusters up to 4 improves the fit; beyond 5 the score falls sharply, indicating overlapping or fragmented groups.

3. Decision: proceed with four clusters for profiling, as this is the best trade-off between cohesion (within-cluster similarity) and separation (between-cluster distance).

4. The implication is, there is meaningful—but not razor-sharp—structure in the data; four behavioural-financial archetypes can be described in the next step.










#### Attach best-K labels and profile clusters

```python
# Code 4-E
kmeans = KMeans(n_clusters=best_k, n_init='auto', random_state=42)
merged['cluster_km'] = kmeans.fit_predict(X_all_scaled)

display(
    merged.groupby('cluster_km')
          [FEATURES_PORT + FEATURES_TRAITS]
          .median()
          .style.format('{:.2f}')
)
```

### Mahalanobis multivariate outliers

#### Compute Mahalanobis distance in combined space

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
from scipy.spatial.distance import mahalanobis
```

```python
# Code 5-A
cov_inv = np.linalg.inv(np.cov(X_all_scaled, rowvar=False))
mean_vec = X_all_scaled.mean(axis=0)

def maha(row):
    return mahalanobis(row, mean_vec, cov_inv)

merged['maha_dist'] = [maha(x) for x in X_all_scaled]
merged['maha_outlier'] = merged['maha_dist'] > 3  # ~99% ellipse

print("Mahalanobis outliers:", merged['maha_outlier'].sum())
```

Outcome:

1. We measured each investor’s overall distance from the centre of the 10-D trait + portfolio space using Mahalanobis distance.

2. Using the conventional |D| > 3 threshold (outside the ≈ 99 % confidence ellipsoid) flags 140 investors—about 47 % of the sample—as multivariate outliers.

3. The unusually large count signals that the combined feature distribution is heavily heavy-tailed; Mahalanobis alone is therefore a blunt instrument here.

4. Next step: intersect these flags with Isolation Forest and LOF results to build a consensus anomaly list, reducing false positives and focusing on the truly rare behaviours.

### Isolation Forest (IF)
#### Density-based anomalies with Isolation Forest

```python
# Code 5-B
iso = IsolationForest(contamination=0.05, random_state=42)
merged['iso_outlier'] = iso.fit_predict(X_all_scaled) == -1
print("IsolationForest outliers:", merged['iso_outlier'].sum())
```

Outcome:

1. Isolation Forest, which detects points that sit in low-density regions of the feature space, identified a much smaller anomaly set.

2. Contamination target 5 % → model returned 15 investors (≈ 5 % of 297) as outliers.

3. This refined list is an order-of-magnitude smaller than the 140 Mahalanobis flags, underscoring how density-based methods filter out heavy-tail but still common behaviours.

4. Next step: combine these 15 with the LOF results and Mahalanobis flags to create a consensus list of truly exceptional investors for the “Outlier Gallery.”

### Local Outlier Factor (LOF)
#### LOF: neighbourhood anomalies

```python
# Code 5-C
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
merged['lof_outlier'] = lof.fit_predict(X_all_scaled) == -1
print("LOF outliers:", merged['lof_outlier'].sum())
```

Outcome:

1. Local Outlier Factor (LOF) flags observations that are far from their nearest neighbours rather than from the global data cloud.

2. With contamination set to 5 % and 20 neighbours, LOF also marks 15 investors as local anomalies—exactly the same count as Isolation Forest but not necessarily the same individuals.

3. Because LOF focuses on neighbourhood density, these 15 investors live in sparse local pockets of the feature space, reinforcing the Isolation Forest result that a small, consistent subset of investors behave very differently from their peers.

4. Next step: intersect the three methods (Mahalanobis, IF, LOF). Investors flagged by at least two techniques will form the “consensus anomaly” list for the Outlier Gallery.

### Consensus anomaly flag & review
#### Who is flagged by at least two methods?

```python
# Code 5-D
merged['outlier_votes'] = (
    merged[['maha_outlier', 'iso_outlier', 'lof_outlier']]
    .sum(axis=1)
)
final_outliers = merged.query('outlier_votes >= 2')

print("Consensus anomalies:", len(final_outliers))
display(final_outliers[
    ['personality_id', 'outlier_votes', 'maha_dist',
     'portfolio_value', 'confidence', 'impulsivity',
     'risk_tolerance']
])
```

Outcome:

1. 18 investors (≈ 6 % of the sample) meet that criterion.

2. 5 investors receive the maximum 3 votes (flagged by all detectors), indicating extreme deviation in both global distance and local density.

3. These investors stand out for different reasons:

a. ID 14 & ID 280 combine very large portfolios (£136 k and £148 k) with Mahalanobis distances > 6, marking them as high-value outliers.

b. ID 200 shows the highest impulsivity (0.91) yet holds a sizeable £44 k portfolio—an unusual risk profile.

c. ID 176 holds only £1 k but ranks in the top quartile for both confidence (0.77) and risk-tolerance (0.72)—a behavioural outlier rather than a financial one.

4. The implication is, these 18 cases merit targeted review—either risk oversight (large concentrated positions) or behavioural coaching (mismatched traits vs. portfolio size). They form the “Outlier Gallery” appendix in the final report.

#### Computing Cook’s D to tag high-influence rows

```python
# Code 5-E
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import OLSInfluence

# 1. Fit the baseline regression
X = sm.add_constant(merged['confidence'])
model = sm.OLS(merged['portfolio_value'], X).fit()
infl  = OLSInfluence(model)

merged['cooks_d'] = infl.cooks_distance[0]

# Common rule-of-thumb threshold = 4 / n
cook_thresh = 4 / len(merged)
hi_influencers = merged.query('cooks_d > @cook_thresh')

print(f"Cook’s-D threshold: {cook_thresh:.4f}")
print(f"High-influence investors: {len(hi_influencers)}")
```

Outcome:

1. We gauged which individual investors exert disproportionate leverage on the confidence → portfolio-value regression.

2. Cook’s-D threshold: 4 / n ≈ 0.0135

3. 15 investors exceed that cut-off and are therefore classed as “high-influence” points.

These rows merit closer inspection and, if needed, a robustness check that re-estimates the regression without them.

4. The implication is, just 5 % of the sample materially sway the fitted slope; dropping them will let us test whether headline relationships are driven by a handful of extreme cases.

### Finance-flavoured metrics helpers

```python
# Code 6-Setup
import numpy as np
import pandas as pd

# Herfindahl-Hirschman Index (HHI) for allocations 
def hhi(values):
    shares = values / values.sum()
    return (shares ** 2).sum()

# Gini coefficient 
def gini(values):
    sorted_x = np.sort(values)
    n = len(values)
    cumx = np.cumsum(sorted_x)
    return (n + 1 - 2 * cumx.sum() / cumx[-1]) / n

# portfolio Value-at-Risk (VaR) at 5 % 
def var_5(values):
    return np.percentile(values, 5)

# build per-person allocation matrix for Section 6 
alloc_matrix = (
    assets_df
    .pivot_table(index='asset_id',
                 columns='asset_allocation',
                 values='asset_value',
                 aggfunc='sum',
                 fill_value=0)
)

# ensure order matches merged index
alloc_matrix = alloc_matrix.reindex(merged['personality_id']).fillna(0)
```

### HHI: How concentrated are portfolios across asset types?

```python
# Code 6-A
merged['HHI_alloc'] = alloc_matrix.apply(hhi, axis=1)
sns.histplot(merged['HHI_alloc'], bins=20)
plt.title('Distribution of portfolio concentration (HHI)')
plt.show()
```

#### Bootstrap 95 % CI for ρ(confidence, portfolio_value)

```python
# Code 8-C
np.random.seed(42)
boot_rho = []
x = merged['confidence'].values
y = merged['portfolio_value'].values

for _ in range(2000):
    idx = np.random.choice(len(x), len(x), replace=True)
    rho, _ = stats.spearmanr(x[idx], y[idx])
    boot_rho.append(rho)

ci_low, ci_high = np.percentile(boot_rho, [2.5, 97.5])
print(f"Bootstrap 95 % CI for ρ: [{ci_low:.3f}, {ci_high:.3f}]")
sns.histplot(boot_rho, kde=True)
plt.axvline(ci_low, color='red', ls='--')
plt.axvline(ci_high, color='red', ls='--')
plt.title('Bootstrap distribution of Spearman ρ')
plt.show()
```
